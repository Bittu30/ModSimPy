```python
bikeshare.olin-=1
bikeshare.wellesley+=1
```

```python
def bike_to_wellesley():
    print('Moving bike to wellesley')
    bikeshare.olin-=1
    bikeshare.wellesley+=1
```

```python
bike_to_wellesley()
```

```python
def bike_to_olin():
    print('Move bike to olin')
    bikeshare.olin+=1
    bikeshare.wellesley-=1
```

```python
bike_to_olin()
```

```python
flip(0.7)
```

```python
if flip(0.5):
    print('heads')
else:
    print('tails')
```

```python
if flip(0.5):
    bike_to_wellesley()
```

```python
if flip(0.33):
    bike_to_olin()
```

```python
def step():
    if flip(0.5):
        bike_to_wellesley()
    if flip(0.33):
        bike_to_olin()
```

```python
step()
```

```python
def step(p1,p2):
    if flip(p1):
        bike_to_wellesley()
    if flip(p2):
        bike_to_olin()
```

```python
step(0.5,0.33)
```

```python
for i in range(4):
    bike_to_wellesley()
```

```python
results=TimeSeries()
```

```python
results[0]=bikeshare.olin
```

```python
for i in range(100):
    step(0.3,0.2)
    results[i]=bikeshare.olin
```

```python
results.mean()
```

```python
results.plot()
```

```python
decorate(title='Olin-Wellesley Bikeshare',
xlabel='Time step (min)',
ylabel='Number of bikes')
results.plot();
```
