```python
def run_simulation(state,p1,p2,num_steps):
    """Sumulate the given number of time steps.
        state:State object
        p1:prob of an olin>welleseley customer arrival
        p2:prob of a welleseley>olin in customer arrival
        num_steps:number of time steps
    """
    
    results=TimeSeries()
    for i in range(num_steps):
        step(state,p1,p2)
        results[i]=state.olin
    plot(results,label='Olin')
```

```python
def bike_to_olin(state):
    """Move one bike from wellersley to Olin.
        state:State Object
    """
    # We decrease one state variable and increase the
    #other, so the total number of bikes is unchanged.
    state.wellesley-=1
    state.olin+=1
```
